```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Membuat objek TfidfVectorizer untuk mengonversi dataset teks menjadi representasi TF-IDF
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(dataset_feature)
```

```python
# Mendapatkan nama fitur yang diekstraksi
features = vectorizer.get_feature_names_out()
```

```python
features
```

```python
# Menghitung nilai IDF untuk setiap kata
idfValues = vectorizer.idf_
```

```python
d = dict(zip(features, 9 - idfValues))  # Memastikan perhitungan IDF tidak bernilai nol
```

```python
sortedDict = sorted(d.items(), key=lambda d: d[1], reverse=True)
```

```python
# Menampilkan 3867 kata dengan nilai IDF tertinggi
for i in range(3867):
    print(sortedDict[i])
```

```python
# Membuat plot WordCloud berdasarkan frekuensi kata-kata
from wordcloud import WordCloud
import matplotlib.pyplot as plot
```

```python
def PlotWordCloud(frequency):
    wordcloudPlot = WordCloud(background_color="white", width=1500, height=1000)
    wordcloudPlot.generate_from_frequencies(frequencies=frequency)
    plt.figure(figsize=(15,10))
    plt.imshow(wordcloudPlot, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
# Memanggil fungsi PlotWordCloud dengan frekuensi kata-kata
PlotWordCloud(d)
```
